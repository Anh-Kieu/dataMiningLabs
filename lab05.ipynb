#### 15) [M] There are many ways to define what we mean by an outlier. A very common technique is known as the IQR rule for outliers. What is a definition of an outlier in terms of IQR?

Outlier: either < Q1  1.5  IQR or > Q3 + 1.5  IQR with Q1 being quartile 1 and Q3 being quartile 3.

#### 16) [P] Write a function called IQR_outlier_limits that takes a dataframe as input, and computes the minimum and maximum outlier thresholds for each variable (column) in the dataframe. The result that is returned will be stored in a data frame.

```python
def IQR_outlier_limit(df):
    lower = df.quantile(0.25) - 1.5 * (df.quantile(0.75) - df.quantile(0.25))
    upper = df.quantile(0.75) + 1.5 * (df.quantile(0.75) - df.quantile(0.25))
    outlier = pd.DataFrame({"min_out": lower, "max_out": upper})
    return outlier
```

#### 17) [P] Show the result of your IQR_outlier_limits function on df_uniform and df_normal.

```python
normal_out = IQR_outlier_limit(df_normal)
normal_out
```

```python
uniform_out = IQR_outlier_limit(df_uniform)
uniform_out
```
